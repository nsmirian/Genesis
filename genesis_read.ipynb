```python
import os
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
import random
with open("mod.out", 'r') as file:
    lines = file.readlines()


wavelength=2e-6
nslice=22
n=0
zstop=1.8
priodlength=0.33
nperiod=zstop/priodlength;
power=np.zeros((nslice, int(nperiod)+1))
for j in range(np.size(lines)-1):
    if lines[j].split()!=[]:
        if lines[j].split()[0]=='power':
            #print(j)
    
            for i in range(int(nperiod)+1):
                tem=lines[j+i+1].split()
                power[n,i]=tem[0]
            
            n +=1
file.close()
#np.arange(10)
s=(np.arange(nslice))*wavelength-3*wavelength
seed= np.stack((s, power[:,-1]), axis=1)
fil2=open('seed2.dat','w')
np.savetxt('seed2.dat', seed, fmt=['%1.14e', '%1.14e'], delimiter='  ')

fil2.writelines('? VERSION=1.0\n')
fil2.writelines('? SIZE=22\n')
fil2.writelines('? COLUMNS ZPOS PRAD0\n')

fil2.close()         
```
